```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import multivariate_normal

```


### Sine Wave with Slope Visualization

This script generates a plot of a sine wave with an added slope,
overlaid with scattered points to simulate noisy data.

The plot demonstrates:
1. A sine wave with amplitude 2 and frequency 3
2. A linear slope of 3
3. 1000 scattered points with random noise added to the sine wave

Dependencies:
- numpy
- matplotlib



```python
# Generate x values
x = np.linspace(0, 10, 1000)

# Generate y values (sine wave with slope)
y = np.sin(3*x)*2 + 3 * x

# Add some random noise to create scattered points
noise = np.random.normal(0, 3, 1000)
y_scattered = y + noise

# Create the plot
plt.figure(figsize=(12, 6))
plt.scatter(x, y_scattered, alpha=0.5)
plt.plot(x, y, color='red', linewidth=2)
plt.title('1000 Random Dots on a Sine Wave with Slope 3')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
# Generate time values (x-axis)
time = np.linspace(0, 10, 1000)

# Generate y values (sine wave with slope)
y = np.sin(3*time)*2 + 3 * time

# Add some random noise to create scattered points
noise = np.random.normal(0, 3, 1000)
y_scattered = y + noise

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(time, y, color='red', linewidth=2, label='Original signal')
plt.plot(time, y_scattered, color='blue', alpha=0.5, label='Noisy samples')
plt.title('Time Series: Sine Wave with Slope and Noisy Samples')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

### Time Series Simulation: Sine Wave with Cumulative Noise

This script generates a time series plot that combines a deterministic signal
(a sine wave with a linear trend) and a stochastic component (cumulative random noise).
This combination creates a pattern reminiscent of stock market behavior.

The plot demonstrates:
1. A sine wave with amplitude 2, frequency 3, and a linear slope of 3
2. Cumulative random noise to simulate unpredictable market movements
3. 1000 time points to represent a detailed time series

Key components:
- Original signal: y = 2 * sin(3t) + 3t
- Noise: Normally distributed random values
- Cumulative noise: Running sum of the noise, simulating persistent market effects

Dependencies:
- numpy
- matplotlib

```python
# Generate time values (x-axis)
time = np.linspace(0, 10, 1000)

# Generate y values (sine wave with slope)
y = np.sin(3*time)*2 + 3 * time

# Add some random noise to create scattered points
noise = np.random.normal(0, 0.3, 1000)  # Increased noise level
y_scattered = y + np.cumsum(noise)

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(time, y, color='red', linewidth=2, label='Original signal')
plt.plot(time, y_scattered, color='blue', alpha=0.5, label='Noisy samples')
plt.title('Time Series: Sine Wave with Slope and Cumulative Noise')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Parameters for the multivariate normal distribution
mean = [0, 0]  # Mean vector
cov = [[1, 0.5], [-0.5, 2]]  # Covariance matrix

# Generate 1000 samples
samples = multivariate_normal.rvs(mean, cov, size=1000)

# Plot the samples
plt.figure(figsize=(10, 10))
plt.scatter(samples[:, 0], samples[:, 1], alpha=0.5)
plt.title('1000 Samples from Multivariate Normal Distribution')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.axis('equal')
plt.show()

# You can also plot the marginal distributions
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.hist(samples[:, 0], bins=30, density=True)
plt.title('Marginal Distribution of X')
plt.subplot(122)
plt.hist(samples[:, 1], bins=30, density=True)
plt.title('Marginal Distribution of Y')
plt.tight_layout()
plt.show()
```

```python
# Parameters for the multivariate normal distribution
true_mean = [0, 0]  # Mean vector
true_cov = [[1, 0.5], [-0.5, 2]]  # Covariance matrix

# Generate 1000 samples
samples = multivariate_normal.rvs(true_mean, true_cov, size=1000)

# Calculate sample mean manually
sample_mean = [sum(samples[:, i]) / len(samples) for i in range(2)]

# Calculate sample covariance manually
n = len(samples)
centered_samples = samples - sample_mean
sample_cov = np.dot(centered_samples.T, centered_samples) / (n - 1)

print("Sample Mean:")
print(sample_mean)
print("\nSample Covariance:")
print(sample_cov)

# Compare with true values
print("\nTrue Mean:")
print(true_mean)
print("\nTrue Covariance:")
print(true_cov)

# Plot the samples
plt.figure(figsize=(10, 10))
plt.scatter(samples[:, 0], samples[:, 1], alpha=0.5)
plt.title('1000 Samples from Multivariate Normal Distribution')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.axis('equal')
plt.show()
```
